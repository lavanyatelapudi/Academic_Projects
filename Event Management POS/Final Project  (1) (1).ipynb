## Class Person and Customer:

* Has customer name
* Has order date of customer
* Has cutomer phone number

```python
import names

class Person(object):
    def __init__(self, first_name, last_name):
        self.first_name = names.get_first_name()
        self.last_name = names.get_last_name()
        
    def full_name(self):
        return f"{self.first_name} {self.last_name}"

class Customer(Person):
    def __init__(self, first_name=None, last_name=None):
        super().__init__(first_name, last_name)
        
    def order_date(self): 
        self.min_year=2022
        self.max_year=datetime.datetime.now().year
        self.start = datetime.datetime(self.min_year, 1, 1, 00, 00, 00)
        self.years = self.max_year - self.min_year + 1
        self.end = self.start + timedelta(days=365 * self.years)
        return ((self.start + (self.end - self.start) * random.random()).strftime("%m/%d/%Y"))
        
    def phone_number(self):
        return (str('+1')+str(randint(100,999))+str('-')+str(randint(100,999))+str('-')+str(randint(1000,9999)))
```

```python
def mergeSort(array,orders):
    '''Use merge sorting to sort orders by date and order amount'''
    if len(array) > 1:
        mid = len(array)//2
        left = array[:mid]
        ordersL = orders[:mid]
        right = array[mid:]
        ordersR = orders[mid:]
        mergeSort(left, ordersL)
        mergeSort(right, ordersR)
 
        i = j = k = 0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                array[k] = left[i]
                orders[k] = ordersL[i]
                i += 1
            else:
                array[k] = right[j]
                orders[k] = ordersR[j]
                j += 1
            k += 1
        while i < len(left):
            array[k] = left[i]
            orders[k] = ordersL[i]
            i += 1
            k += 1
        while j < len(right):
            array[k] = right[j]
            orders[k] = ordersR[j]
            j += 1
            k += 1
```

```python
def enum_list(l):
     for i,j in enumerate(l,1):
        print(f"({i}). {j}")
```

```python
def sort_by_date_or_amount(orders):
    '''Sort order by date and order amount using Binary Search and Merge sort'''
    print('Sort order by order amount: Press 1\nSort order by date: Press 2')
    try:
        opt = int(input('Enter your choice: '))
        if opt == 1:
            by_amount = [i[2] for i in orders]
            mergeSort(by_amount, orders)
            print('\n\n\033[1mOrders sorted by amount\033[0m\n')
            enum_list(orders)
        elif opt == 2:
            by_date = [datetime.datetime.strptime(i[4], '%m/%d/%Y').date() for i in orders]
            mergeSort(by_date, orders)
            print('\n\n\033[1mOrders sorted by date\033[0m\n')
            enum_list(orders)
        else:
            print('Invalid choice. Please enter either 1 or 2.')
    except ValueError:
        print('Invalid input. Please enter a valid integer.')
    except Exception as e:
        print(f'An error occurred: {str(e)}')
```

```python
def binary_search(array,e):
    '''Used binary search because it is fastest.Array is merge sorted before searched'''
    left = 0
    right = len(array) - 1
    
    while (left <= right): 
        middle = (left + right) // 2
        if (array[middle] == e): 
            return middle
        elif (array[middle] < e): 
            left = middle + 1
        else: 
            right = middle - 1
    return 'Not Found'

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1
```

```python
def search_by_name_or_date(orders):
    print('Search order by name: Press 1')
    print('Search order by date: Press 2')
    print('Get orders list before a certain date: Press 3')
    print('Get orders list after a certain date: Press 4')

    user_inp = int(input('Press 1/2/3/4: '))

    if user_inp == 1:
        ordername = [i[0] for i in orders]
        mergeSort(ordername, orders)
        give_name = str(input('Enter customer name (Firstname Lastname): '))
        index = binary_search(ordername, give_name)
        if index != -1:
            print(orders[index])
        else:
            print('Order not found.')

    else:
        orderdate = [i[4] for i in orders]
        mergeSort(orderdate, orders)

        if user_inp == 2:
            give_date = str(input('Enter order date (mm/dd/yyyy): '))
            index = binary_search(orderdate, give_date)
            if index != -1:
                print(orders[index])
            else:
                print('Order not found.')

        elif user_inp == 3:
            prior_date = str(input('Enter order date (mm/dd/yyyy): '))
            index = binary_search(orderdate, prior_date)
            if index != -1:
                print(orders[:index])
            else:
                print('Order not found.')

        elif user_inp == 4:
            after_date = str(input('Enter order date (mm/dd/yyyy): '))
            index = binary_search(orderdate, after_date)
            if index != -1:
                print(orders[index + 1:])
            else:
                print('Order not found.')
```


##### Time Complexity for Sorting:
Time Complexity It describes the total time taken to run each line of an algorithm.

* sort_date_by_amount(orders) function involves a only a for loop and mergeSort method.
* Time complexity for mergeSort is same for worst case(Big-O),best case(Big-Omega) and average case(Big-Theta) i.e. O(n*logn)
Time complexity for sorting, T(n)=O(n)+ O(nlogn) = O(nlogn)

##### Time Complexity for Searching:

search_by_name_or_date(orders) has mergeSort function called, binary search algorithm called and has only a for loop everytime user prompts
* Time complexity for Binary Search = O(logn)
* Time complexity for Merge Sort = O(n*logn)
* Time complexity for for loop = O(n)
* Time complexity for searching is, T(n)=O(logn)+O(nlogn)+O(n)=O(nlogn)


##### Space Complexity for Sorting:
**Space Complexity It describes the amount of memory an algorithm takes in terms of the amount of input to the algorithm.**

sort_date_by_amount(orders) function involves a only a for loop and mergeSort method.
* Space complexity for 'for' loop is O(1)
* Space complexity for merge sort is O(n)
* Space complexity for sorting is, O(1)+O(n)=O(n)
##### Space Complexity for Searching:

search_by_name_or_date(orders) has mergeSort function called, binary search algorithm called and has only a for loop.
* Space complexity for 'for' loop is O(1)
* Space complexity for merge sort is O(n)
* Space complexity for binary search is O(1)
* Space complexity for searching, O(1)+O(n)+O(1)=O(n)
